````python
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
````

````python
class calc:
    @staticmethod
    def score(pred):
        if type(pred)==str:
            return 999
        else:
            target = pd.read_csv(path+"평가지표데이터.csv")
            value = 0
            for i in pred.columns.unique()[1:]:
                rmse =  (mean_squared_error(target[[i]], pred[[i]]))**0.5
                r2 = r2_score(target[[i]], pred[[i]])
                value += rmse/r2
            print(f"점수: {value/4:.4f}")
            return value/4

    @staticmethod
    def score2(pred):
        if type(pred)==str:
            return 999
        else:
            target = pd.read_csv(path2+"평가지표데이터.csv")
            value = 0
            for i in pred.columns.unique()[1:]:
                rmse =  (mean_squared_error(target[[i]], pred[[i]]))**0.5
                r2 = r2_score(target[[i]], pred[[i]])
                value += rmse/r2
            print(f"점수: {value/4:.4f}")
            return value/4
````
