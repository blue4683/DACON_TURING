```python
import datetime
import requests
import json
import pandas as pd

# tide_level
def 강화대교_조위_sub_request(date):
    URL = f'https://www.khoa.go.kr/api/oceangrid/tideObs/search.do?ServiceKey=kH9DXri8W9Ny6w0onLeuqw==&ObsCode=DT_0032&Date={date.strftime("%Y%m%d")}&ResultType=json'
    print(URL)
    request_headers = { 
    'User-Agent' : ('Mozilla/5.0 (Windows NT 10.0;Win64; x64)\
    AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98\
    Safari/537.36'), } 
    response = requests.get(URL, headers = request_headers)
    return response.json()
```

```python
temp = []

def 강화대교_조위(startdate, enddate):
    date = startdate

    while(date!=enddate):
        강화대교_조위_result = 강화대교_조위_sub_request(datetime.datetime.strptime(date, "%Y%m%d"))
        temp.append([(i['record_time'], i['tide_level']) for i in 강화대교_조위_result['result']['data']])
        date = datetime.datetime.strptime(date, "%Y%m%d") + datetime.timedelta(days=1)
        date = date.strftime("%Y%m%d")
        print(date)
    return temp
    
a = 강화대교_조위("20220719", "202200830")
```

```python
df = pd.DataFrame()
for i in range(len(a[1:])+1):
    if len(df)==0:
        df = pd.DataFrame(a[i])
    else:
        df = pd.concat([df, pd.DataFrame(a[i])], axis=0)
df
```
